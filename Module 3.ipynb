## Part 3
What would have happened to prices had we increased the supply of shoes available? What is the intuition for that?
### Simulated Scenario:
- Original supply: 40 pairs of slides
- Increased supply scenarios: 60 pairs, 80 pairs, 100 pairs
- Assume demand decreases linearly with price increase above the original clearing price
- Original clearing price: $200
- Assume the demand elasticity is such that every 10 additional pairs of slides require a 5% decrease in price to clear the market.

```python
# Original assumptions
original_supply = 40
clearing_price_original = 200

# Increased supply scenarios
increased_supplies = [60, 80, 100]

# Assuming a 5% price decrease is needed for every 10 additional pairs to clear the market
price_decreases = [(supply - original_supply) / 10 * 0.05 for supply in increased_supplies]
clearing_prices_new = [clearing_price_original * (1 - decrease) for decrease in price_decreases]

# Resulting clearing prices for increased supplies
supply_price_changes = dict(zip(increased_supplies, clearing_prices_new))
supply_price_changes
```

# Analysis of Increased Supply on Prices

Simulating the impact of increased supply on the clearing prices under the assumption that demand decreases linearly with price increases, we find:

- With a supply of **60 pairs**: The clearing price would adjust to **$180**, a decrease from the original $200.
- With a supply of **80 pairs**: The clearing price further decreases to **$160**.
- With a supply of **100 pairs**: The clearing price drops to **$140**.

## Intuition Behind the Price Changes:

The simulated decrease in clearing prices with increased supply follows the basic economic principle of supply and demand:

- **Increased Supply:** Offering more pairs of the exclusive slides increases the market's supply.
- **Price Adjustment:** To ensure all additional pairs are sold, the clearing price must decrease to stimulate demand. This is because the marginal buyer’s willingness to pay decreases as the availability of the product increases.
- **Market Equilibrium:** The new clearing prices represent points where market supply and demand meet under the increased supply conditions.

## Conclusion:

Increasing the supply of shoes in the Ben Baller IPO auction would likely have led to a decrease in the clearing price, assuming demand remained constant or decreased linearly with price increases. This is consistent with the intuition that making a product less scarce (more available) reduces its exclusivity and, therefore, the amount buyers are willing to pay. The extent of the price change would depend on the actual demand elasticity, which this simulation approximates linearly for illustrative purposes.

## Part 5
Are there any outlier bids in the data? For which shoe sizes and colors? How should you treat outliers?



```python
# Calculate the IQR and identify outliers for each shoe size and color combination
def identify_outliers(group):
    q1 = group['amount'].quantile(0.25)
    q3 = group['amount'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = group[(group['amount'] < lower_bound) | (group['amount'] > upper_bound)]
    return outliers

# Apply the function to each group
outlier_bids = ipo_bids.groupby(['product_type', 'shoe_size']).apply(identify_outliers).reset_index(drop=True)

# Displaying the outliers identified
outlier_bids[['product_type', 'shoe_size', 'amount']]
```

# Outlier Bids Analysis:

The analysis identified outlier bids across various shoe sizes and colors in the dataset. Here are some observations:

- Outliers are present in both black and red shoes, indicating that bids significantly higher or lower than the typical bid range were placed across different sizes.
- The outlier amounts vary widely, with some bids as low as $200 and others reaching up to $100,000,000, suggesting significant variability in bidder behavior or potential data entry errors.
- Outliers were detected across a range of shoe sizes, indicating that unusual bidding was not confined to a particular size or color.

## How to Treat Outliers:

The treatment of outliers should be carefully considered based on the context of the analysis and the goals of the study. Here are some general approaches:

- **Exclusion:** Remove outliers from the dataset to focus the analysis on the typical bid range. This approach is suitable when outliers are considered noise or errors.

- **Further Investigation:** Examine outliers more closely to understand their cause. If outliers represent genuine bids, they might offer insights into particular market segments or bidder behavior.

- **Capping:** Replace outlier values with the nearest non-outlier value (e.g., the upper or lower bounds of the typical bid range). This approach retains the outliers in the analysis but reduces their impact.

- **Separate Analysis:** Conduct a separate analysis for outliers to explore patterns or behaviors among extreme bids without affecting the overall analysis.

- **Statistical Adjustment:** Use robust statistical methods that are less sensitive to outliers to analyze the data.

## Conclusion:

Identifying outliers in the bid data reveals significant variations in bidding behavior for different shoe sizes and colors. The choice of how to treat these outliers should align with the specific objectives of the analysis and the nature of the data. Whether outliers are removed, adjusted, or analyzed separately, the rationale behind these decisions should be clearly documented and justified based on the research questions or business objectives at hand.
